```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


target_variable = 'Classification'

correlations = df.corrwith(df[target_variable])

abs_correlations = correlations.abs()

correlation_df = pd.DataFrame({'Feature': abs_correlations.index, 'Correlation': abs_correlations.values})

correlation_df_sorted = correlation_df.sort_values(by='Correlation', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(data=correlation_df_sorted, x='Correlation', y='Feature', palette='coolwarm')
plt.xlabel('Absolute Correlation with Target Variable')
plt.ylabel('Feature')
plt.title('Absolute Correlation Matrix of Features with Target Variable')
plt.show()
```

# Thresholding Features

```python
threshold = 0.6

columns_to_drop = correlation_df_sorted[correlation_df_sorted['Correlation'] < threshold]['Feature']

df.drop(columns=columns_to_drop, inplace=True)
columns_to_drop
```

# train-test split

```python
import pandas as pd
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.25, random_state=42)
```

```python
train_df
```

```python
del df
```

# Sampling Classes

```python
t1 = time.time()
groups = train_df.groupby('Classification')

initial_counts = train_df['Classification'].value_counts()

sampled_dfs = []

for _, group in groups:
    if len(group) < 200000:
        num_duplicates = 200000 // len(group)
        remainder = 200000 % len(group)
        sampled_df = pd.concat([group] * num_duplicates + [group.sample(remainder, replace=False)])
    else:
        sampled_df = group.sample(200000, replace=False, random_state=42)
    sampled_dfs.append(sampled_df)

train_sampled_df = pd.concat(sampled_dfs)

final_counts = train_sampled_df['Classification'].value_counts()

print("Initial counts per class:")
print(initial_counts)
print("\nFinal counts per class after balancing:")
print(final_counts)
print("The time taken is : ",time.time()-t1)
```

```python
del initial_counts, final_counts, train_df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X_train = train_sampled_df.drop(columns=['Classification'])

y_train = train_sampled_df['Classification']

rf_classifier = RandomForestClassifier(n_estimators=200, random_state=42)

rf_classifier.fit(X_train, y_train)

```

```python
del train_sampled_df,X_train, y_train
```

```python
X_test = test_df.drop(columns=['Classification'])
y_test = test_df['Classification']
```

```python
import sys

def get_memory_usage():
    variables = globals().copy()
    memory_usage = {}
    
    for var_name, var in variables.items():
        size = sys.getsizeof(var)
        memory_usage[var_name] = size
    
    return memory_usage

memory_usage = get_memory_usage()

for var_name, size in memory_usage.items():
    print(f"Variable '{var_name}': {size} bytes")
```
